# Creating dummy varibles for genres column

```python
Genre=pd.get_dummies(data[["genre"]])
Genre.tail()
```

```python
Rating=pd.get_dummies(data[["content_rating"]])
Rating.tail()
```

# Concating the relavent columns from data to new data with dummy variables which already created:

```python
new_data=pd.concat([Genre,Rating,data[["star_rating","duration"]]],1)
new_data.tail()
```

```python
new_data.tail()
```

```python
from sklearn.preprocessing import MinMaxScaler
Robust=MinMaxScaler()
New_data=Robust.fit_transform(new_data)
```

```python
np.round(New_data,2)
```

# Using KNN algorithm to find our prediction

```python
reccomended_movie = NearestNeighbors(n_neighbors=10, algorithm='auto').fit(New_data)
```

```python
distances, indices = reccomended_movie.kneighbors(New_data)
```

```python
def Movie_title(name):
    return data[data["title"]==name].index.tolist()[0]
```

```python
Movie_title("Pulp Fiction")
```

```python
def Similar_movie_title(title=None):
     if title:
        found_title = Movie_title(title)
        for id in indices[found_title][1:]:
            print(data.iloc[id]["title"])
```
